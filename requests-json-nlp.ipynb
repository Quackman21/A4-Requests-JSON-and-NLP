2. Read in the contents of your file.  Print the lyrics of the song (not the entire dictionary!) and use spaCyTextBlob to perform sentiment analysis on the lyrics.  Print the polarity score of the sentiment analysis.  Given that the range of the polarity score is `[-1.0,1.0]` which corresponds to how positive or negative the text in question is, do you think the lyrics have a more positive or negative connotaion?  Answer this question in a comment in your code cell.

```python
import json
import lyricsgenius
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob


genius = lyricsgenius.Genius('YOUR_ACCESS_TOKEN')

# Search for the song "Wonderwall" by the artist "Oasis".
song = genius.search_song("Wonderwall", "Oasis")

# Look up song.
if song is not None and song.lyrics is not None:
    # Print the lyrics of the song.
    print(song.lyrics)

    # Save the lyrics to a JSON file.
    lyrics_data = {
        "song_title": song.title,
        "artist": song.artist,
        "lyrics": song.lyrics
    }

    with open("wonderwall_lyrics.json", "w") as json_file:
        json.dump(lyrics_data, json_file, indent=4)

    print("Lyrics saved to 'wonderwall_lyrics.json'.")
else:
    print("Lyrics not found for the song.")

# Read the contents of the JSON file.
with open("wonderwall_lyrics.json", "r") as json_file:
    lyrics_data = json.load(json_file)

# Extract the lyrics from the loaded JSON data.
lyrics = lyrics_data["lyrics"]

# Perform sentiment analysis using spaCyTextBlob.
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("spacytextblob")
doc = nlp(lyrics)

# Print the polarity score of the sentiment analysis.
polarity_score = doc._.polarity
print("Polarity score:", polarity_score)

# Answer to the question:
# Based on the polarity score, if the score is positive (closer to 1.0), the lyrics have a more positive connotation.
# If the score is negative (closer to -1.0), the lyrics have a more negative connotation.
```

3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

```python
import lyricsgenius
import json

def get_and_save_lyrics(artist, song, filename):
    # Replace 'YOUR_ACCESS_TOKEN' with your actual access token or set it as an environment variable.
    genius = lyricsgenius.Genius('YOUR_ACCESS_TOKEN')

    # Search for the specified song by the artist.
    song_result = genius.search_song(song, artist)

    # Check if the song was found and has lyrics.
    if song_result is not None and song_result.lyrics is not None:
        # Extract the lyrics from the song result.
        lyrics = song_result.lyrics

        # Save the lyrics to a JSON file.
        lyrics_data = {
            "song_title": song_result.title,
            "artist": song_result.artist,
            "lyrics": lyrics
        }

        with open(filename, "w") as json_file:
            json.dump(lyrics_data, json_file, indent=4)

        print(f"Lyrics for '{song}' by '{artist}' saved to '{filename}'.")
    else:
        print(f"Lyrics for '{song}' by '{artist}' not found.")

# Example usage:
get_and_save_lyrics("Calvin Harris", "Summer", "Summer_lyrics.json")
get_and_save_lyrics("Calvin Harris", "One Kiss", "One_Kiss_lyrics.json")
get_and_save_lyrics("Red Hot Chili Peppers", "Cant Stop", "Cant_Stop_lyrics.json")
get_and_save_lyrics("The Clash", "London Calling", "London_Calling_lyrics.json")
```

4. Write a function that takes the name of a file that contains song lyrics, loads the file, performs sentiment analysis, and returns the polarity score.  Use this function to print the polarity scores (with the name of the song) of the three files you created in question 3.  Does the reported polarity match your understanding of the song's lyrics? Why or why not do you think that might be?  Answer the questions in either a comment in the code cell or a markdown cell under the code cell.

```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob
import json

def get_polarity_score_from_file(filename):
    with open(filename, "r") as json_file:
        data = json.load(json_file)
        lyrics = data["lyrics"]

    nlp = spacy.load("en_core_web_sm")
    nlp.add_pipe("spacytextblob")
    doc = nlp(lyrics)

    return doc._.polarity

# Example usage to get and print polarity scores for the three files:
files = ["Summer_lyrics.json", "One_Kiss_lyrics.json", "Cant_Stop_lyrics.json"]

for filename in files:
    polarity_score = get_polarity_score_from_file(filename)
    song_title = filename.replace("_lyrics.json", "").replace("_", " ").title()
    print(f"Polarity score for '{song_title}': {polarity_score}")
```
